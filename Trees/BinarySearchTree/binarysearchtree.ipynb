# Binary Search Tree
### Operations: insertion, search, deletion, minimum, maximum

```python
class Node(object):
    def __init__(self,data =None):
        self.left= None
        self.right= None
        self.parent= None
        self.data = data
    def setLeft(self, node):
        self.left = node
    def setRight(self, node):
        self.right = node
    def getLeft(self):
        return self.left
    def getRight(self):
        return self.right
    def setData(self, data):
        self.data = data
    def getData(self):
        return self.data

class Tree(object):
    def __init__(self):
        self.root = None
    
    def insert(self,node):
        newNode= Node(node)
        y=None
        x=self.root
        while(x is not None):
            y=x
            if(newNode.data < x.data):
                x= x.left
            elif(newNode.data > x.data):
                x=x.right
            else:
                print("duplicate value not accepted")
                return
        newNode.parent=y
        if(y == None):
            self.root = newNode
        elif(newNode.data< y.data):
            y.left=newNode
        elif(newNode.data > y.data):
            y.right = newNode
        else:
            print("duplicate value not accepted")
            return
        
    def transplant(self,u,v):
        if(u.parent ==None):
            self.root= v
        elif(u == u.parent.left):
            u.parent.left = v
        else:
            u.parent.right= v
        if(v != None):
            v.parent= u.parent
    
    def delete(self,z):
        if(z.left == None):
            self.transplant(z,z.right)
        elif(z.right == None):   
            self.transplant(z,z.left)
        else:
            y=self.minimum(z.right)
            if(y.parent !=z):
                self.transplant(y,y.right)
                y.right=z.right
                y.right.parent = y
            self.transplant(z,y)
            y.left= z.left
            y.left.parent = y
   
    def treeSuccessor(self,x):
        if(x.right != None):
            
            return self.minimum(x.right)
        y= x.parent
        
        while(y != None and x == y.right):
            x=y
            y=y.parent
            print(y," ")
            
        return y
    
    
    
    def treePredecessor(self,x):
        if(x.left != None):
            return self.maximum(x.left)
        y= x.parent
        while(y != None and x==y.left):
            x=y
            y=y.parent
        return y
    
    #heigh or depth
    def maxHeight(self,root):
        if(root == None):
            return 0
        else:
            lDepth= self.maxHeight(root.left)
            rDepth= self.maxHeight(root.right)
        if(lDepth > rDepth):
            return lDepth+1
        else:
            return rDepth+1
    
    def search(self, x, data):
        if(x== None or data== x.data):
            return x
        if(data < x.data):
            self.search(x.left, data)
        else:
            self.search(x.right, data)
            
    def iterativeSearch(self,x,data):
        while(x != None and data != x.data):
            if(data < x.data):
                x=x.left
            else:
                x= x.right
        if(x== None):
            print("\nNo such value")
            
        elif(data== x.data):
            print("\ndata found")
            
        
        
    def minimum(self, x):
        while(x.left is not None):
            x= x.left
        return x
    
    def maximum(self,x):
        while(x.right is not None):
            x= x.right
        return x
    
    def inOrder(self, x):
        if(x != None):
            self.inOrder(x.left)
            print(x.data, end=" ")
            self.inOrder(x.right)
    
    def preOrder(self, x):
        if(x != None):
            print(x.data, end=" ")
            self.preOrder(x.left)
            self.preOrder(x.right)
    
    def postOrder(self, x):
        if(x != None):
            self.postOrder(x.left)
            self.postOrder(x.right)
            print(x.data, end=" ")
    def lowestCommonAncestor(self,root, n1, n2): 
        if root is None: 
            return None
        if(root.data > n1 and root.data > n2): 
            return self.lowestCommonAncestor(root.left, n1, n2)  
        if(root.data < n1 and root.data < n2): 
            return self.lowestCommonAncestor(root.right, n1, n2) 
        return root 
            

tree = Tree()
tree.insert(50)
tree.insert(30)
tree.insert(20)
#tree.insert(20)
tree.insert(40)
tree.insert(70)
tree.insert(60)
tree.insert(80)

root= tree.root
tree.inOrder(root)
print("\nsearch: ",tree.search(root,50).data)
min= tree.minimum(root).data
print("minimum: ",min)
max= tree.maximum(root).data
print("maximum: ",max)
tree.iterativeSearch(root,100)
#print("root: ",root.data)

successor = tree.treeSuccessor(root).data
print("Successor: ",successor)

predecessor = tree.treePredecessor(root).data
print("Predecessor: ",predecessor)
print("Inorder:")
tree.inOrder(tree.root)
print("\nPre order: ")
tree.preOrder(tree.root)
print("\nPost Order: ")
tree.postOrder(tree.root)
#tree.delete(root.right)
height= tree.maxHeight(root)

print("\nHeight: ",height)

n1,n2=80,60
lca= tree.lowestCommonAncestor(tree.root,n1,n2)
print("Lowest Common Ancestor of ",n1," and ",n2," is :",lca.data)

secondLargest= tree.maximum(root)
print("Second Largest Element",secondLargest.parent.data)
secondSmallest= tree.minimum(root)
print("Second Smallest Element",secondSmallest.parent.data)

  
```
